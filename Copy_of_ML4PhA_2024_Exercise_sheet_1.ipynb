````python
from numpy.random import random
import matplotlib.pyplot as plt

def return_random_sample():
    # Your solution here
    randNum = random()
    #print(randNum)
    if randNum<0.2:
      choice = 0
    elif randNum < 0.5:
      choice = 1
    else:
      choice = 2
    return choice
````

### 2.2 Marginal normal distributions (3 points)

*Topics: multivariate normal & marginal distributions*

Consider a multivariate Gaussian of the form
$$P(\mathbf{x}) = \mathcal{N}(\mathbf{x}|\mathbf{\mu}, \mathbf{\Sigma}) = \frac{1}{(2\pi)^{D/2} |\mathbf{\Sigma}|^{1/2}} \exp\left(-\frac{1}{2} (\mathbf{x}-\boldsymbol{\mu})^{T} \mathbf{\Sigma}^{-1}(\mathbf{x}-\boldsymbol{\mu})\right) \, ,$$

where $\mathbf{x}$ and $\boldsymbol{\mu}$ are vectors with $D$ entries and $\mathbf{\Sigma}$ is a $D \times D$ covariance matrix. $|\mathbf{\Sigma}|$ is the determinant of this matrix.

#### a (2 points)
Consider the case where $D = 2$. Use the following three sets of ($\boldsymbol{\mu}$, $\mathbf{\Sigma}$) to create 2D contour- and 3D surface plots of the corresponding PDFs (with arbitrary contour levels)

$$ \boldsymbol{\mu_1} = \begin{pmatrix} 5 \\ 5 \end{pmatrix} ~ \mathbf{\Sigma_1} = \begin{pmatrix} 1 & 0 \\ 0 & 1 \end{pmatrix} \, ,$$

$$ \boldsymbol{\mu_2} = \begin{pmatrix} 5 \\ 5 \end{pmatrix} ~ \mathbf{\Sigma_2} = \begin{pmatrix} 1 & -0.5 \\ -0.5 & 1 \end{pmatrix} \, ,$$

$$ \boldsymbol{\mu_3} = \begin{pmatrix} 5 \\ 5 \end{pmatrix} ~ \mathbf{\Sigma_3} = \begin{pmatrix} 1 & 0.7 \\ 0.7 & 1 \end{pmatrix} \, .$$

Implement the multivariate Gaussian yourself using numpy matrix multiplication.

````python
import numpy as np
import matplotlib.pyplot as plt

def mvgauss(x, mu, sigma):
    """
    Calculate the multivariate Gaussian density for a given x, mean vector mu, and covariance matrix sigma.
    """
    D = mu.shape[0]  # Dimensionality
    x_m = x - mu
    return (1. / (np.sqrt((2 * np.pi)**D * np.linalg.det(sigma))) *
            np.exp(-0.5 * np.dot(x_m.T, np.dot(np.linalg.inv(sigma), x_m))))

# Put your solution here
def plot_mvgauss(mu, sigma, title):
    """
    Plot the 2D contour and 3D surface for a multivariate Gaussian distribution.

    Parameters:
    - mu: Mean vector of the Gaussian distribution.
    - sigma: Covariance matrix of the Gaussian distribution.
    - title: Title for the plot.
    """
    # Grid of points
    x, y = np.mgrid[0:10:.1, 0:10:.1]
    #print('x=\n',x)
    pos = np.dstack((x, y))

    # Calculate PDF values
    Z = np.zeros(x.shape)
    for i in range(x.shape[0]):
        for j in range(x.shape[1]):
            Z[i, j] = mvgauss(pos[i, j], mu, sigma)

    # 2D Contour Plot
    plt.figure(figsize=(4, 3))
    plt.contour(x, y, Z)
    plt.title(f'2D Contour Plot: {title}')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.show()

    # 3D Surface Plot
    fig = plt.figure(figsize=(7, 6))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(x, y, Z, cmap='viridis')
    ax.set_title(f'3D Surface Plot: {title}')
    ax.set_xlabel('X-axis')
    ax.set_ylabel('Y-axis')
    ax.set_zlabel('PDF Value')
    plt.show()

    return Z

# Now, call the function for each set of parameters
mu1=mu2=mu3 = np.array((5,5))
sigma1 = np.array([[1,0],[0,1]])
sigma2 = np.array([[1,-0.5],[-0.5,1]])
sigma3 = np.array([[1,0.7],[0.7,1]])

Z_Sigma1 = plot_mvgauss(mu1, sigma1, 'PDF with Sigma1')
Z_Sigma2 = plot_mvgauss(mu2, sigma2, 'PDF with Sigma2')
Z_Sigma3 = plot_mvgauss(mu3, sigma3, 'PDF with Sigma3')

````

### 2.3 Inverse problem with Poisson distribution (2 points)

*Topic: Bayes theorem, uniform and Poisson distributions, discrete and continuous random variables*

The number of steps that a robot takes in a given amount of time is $s$, where $s$ is Poisson distributed with mean value $\lambda$. Derive an analytic expression for the posterior of $\lambda$ as a function of the observed number of steps $s$. You can assume a flat (uniform) prior for $\lambda$.

*Put your solution here*


\begin{gather*}
\begin{aligned}
P( \lambda |s) & =\frac{P( s|\lambda ) P( \lambda )}{P( s)}\\
 & =\frac{P( s|\lambda ) P( \lambda )}{\sum P( s_{i} |\lambda ) P( \lambda )}
\end{aligned}\\
\end{gather*}

Since we assume the prior to be uniform, which is a constant, $P(\lambda)$ can be about dropped. So,

\begin{equation*}
\begin{aligned}
P( \lambda |s) & =\frac{P( s|\lambda )}{\int P( s|\lambda ) d\lambda }\\
 & =\frac{\frac{e^{-\lambda } \lambda ^{s}}{s!}}{\int \frac{e^{-\lambda } \lambda ^{s}}{s !} d\lambda }\\
 & =\frac{\frac{e^{-\lambda } \lambda ^{s}}{s!}}{\frac{1}{s!} \Gamma ( s+1)}\\
 & =\frac{\Gamma ( s+1,\lambda )}{\Gamma ( s+1)}
\end{aligned}
\end{equation*}
